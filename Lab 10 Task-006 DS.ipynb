```python
def encode(df, cols):
    for col in cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
    return df

x = ['Home Team', 'Away Team', 'Result']
data = encode(data, x)
```

```python
def fillna(x):
    for i in x:
        data[i] = data[i].fillna(data[i].median())
columns = ["homeScore",'homeXG','awayScore','awayXG','Result']
fillna(columns)
```
