<b> data preprocessing</b>

```python
data.drop('Matchday', axis = 1 , inplace = True)
data.drop('Date', axis = 1 , inplace = True)
data.drop('Time', axis = 1 , inplace = True)
data.drop('Referee', axis = 1 , inplace = True)
data.drop('Stadium', axis = 1 , inplace = True)
data.drop('Attendance', axis = 1 , inplace = True)
data.drop('*Additional Stats', axis = 1 , inplace = True)
```

```python
def floattoint(x):
    for i  in x:
        data[i] = data[i].astype('int64') 
x = ['homeScore','homeXG', 'awayScore', 'awayXG']
floattoint(x)
data.info()
```

```python
def encode(df, cols):
    for col in cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
    return df

x = ['Home Team', 'Away Team', 'Result']
data = encode(data, x)
```

```python
def fillna(x):
    for i in x:
        data[i] = data[i].fillna(data[i].median())
columns = ["homeScore",'homeXG','awayScore','awayXG','Result']
fillna(columns)
```

```python
data.info()
```

<b> Data Train Test Splitting </b>

```python
X = data.drop('Result', axis=1)
y = data['Result']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y , test_size=0.2, shuffle = False)

model = RandomForestClassifier()
model.fit(Xtrain, ytrain)
```

```python
rfc = model.predict(Xtest)
print(rfc)
```

```python
accuracy = accuracy_score(ytest, rfc)
print(round(accuracy, 3))
```
